# Basic implementation using simple functions

```python
# importing libraries
import pandas as pd
import warnings

warnings.filterwarnings('ignore')

# reading data
df = pd.read_csv('sim_lin_reg.csv')

# mathematical background of the model
sigma_x = df['x'].sum()
sigma_y = df['y'].sum()

sigma_xy = sum([df['x'][i]*df['y'][i] for i in range(len(df['x']))])

sigma_x2 = sum([df['x'][i]*df['x'][i] for i in range(len(df['x']))])

slope = (len(df['x'])*sigma_xy-sigma_x*sigma_y)/(len(df['x'])*sigma_x2-(sigma_x)**2)
intercept = (sigma_y-slope*sigma_x)/len(df['x'])
```

```python
# printing slope and intercept calculated
print("Slope:",slope,"\nIntercept:",intercept)
```

```python
# user input
predictor = float(input("Enter value to be predicted on: "))

# predicting on above inputs
prediction = slope*predictor+intercept

# displaying results
print("Predicting on:",predictor,"\tValue predicted:",prediction)
```

```python
# plotwork

import numpy as np
import matplotlib.pyplot as plt

plot_x = np.linspace(0, 10, 100)

plot_y = slope * plot_x + intercept

plt.figure(figsize=(8, 6))
plt.plot(plot_x, plot_y, label=f'y = {slope}x + {intercept}', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot with Given Slope and Intercept')
plt.legend()
plt.grid(True)
plt.show()
```

# implemention using Numpy

```python
# turning dataframe into numpy ndarray
df = df.to_numpy().T

# mathematical background of the model
sigma_x = np.sum(df[0])
sigma_y = np.sum(df[1])

sigma_xy = np.sum(np.multiply(df[0],df[1]))

sigma_x2 = np.sum(np.square(df[0]))

slope = (df[0].size*sigma_xy-sigma_x*sigma_y)/(df[0].size*sigma_x2-pow(sigma_x,2))
intercept = (sigma_y-slope*sigma_x)/(df[0].size)

# printing slope and intercept calculated
print("Slope:",slope,"\nIntercept:",intercept)

# user input
predictor = float(input("Enter value to be predicted on: "))

# predicting on above inputs
prediction = slope*predictor+intercept

# displaying results
print("Predicting on:",predictor,"\tValue predicted:",prediction)
```
